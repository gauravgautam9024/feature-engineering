```python
c = pd.read_csv(r"C:\Users\Lenovo\Desktop\gaurav\covid_toy.csv")
c
```

```python
"""
Feature engineering are provide some method and encoding techniques:
1.
Label Encoding: Label Encoding is provide every category to unique integer and assign . And its also use in  data  ordinality made introduce, its does not desirable.
Label Encoding:
Lable Encoder()
Make: Toyota=0, Honda=1, Ford=2 """
label = LabelEncoder()
c["gender_encoding"] = label.fit_transform(c["gender"])
c
```

```python
"""
2.
One-Hot Encoding: Its provide every category to binary column and create it . Har observation ko uske category ke corresponding column mein 1 milta hai aur sabhi dusre columns mein 0 milte hain.
One-Hot Encoding:

For "Make":

Toyota Honda Ford

1 0 0"""
one_hot_encoder = pd.get_dummies(c["cough"],prefix ="cough")
one_hot_encoder
# concat 
v  =pd.concat([c,one_hot_encoder],axis =1)
v
```

```python
"""
3. **Ordinal Encoding**:
Its also like label encoding, lekin yeh categories ke order ko consider karta hai aur integers ko uss order ke hisaab se assign karta hai. 
For example, "low," "medium," aur "high" ko 0, 1, aur 2 ke taur par encode kiya ja sakta hai.
"""
ordinal_mapping_make = {"Mild": "M", "Strong": "s"}
d = c['cough_mapping'] = c['cough'].map(ordinal_mapping_make)
d
```

```python
# Create a simple dataset  with skewed distribution
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(0)
data ={ 'income': np.random.exponential(scale=1000,size=1000)
      }
df = pd.DataFrame(data)
df.head(30)
sns.histplot(df['income'],kde=True,edgecolor="black")
plt.xlabel(" distribution of income (before transformation)")
plt.title("income number")
plt.ylabel("frequency")
plt.show()
```

```python
#apply log transformation to income feature
df["Income_log"]= np.log(df["income"])           # np.log is a np function that calculates the natural logaritham(base e) number of every element
df.head(10)                                       # log are also use to distribute number as equal parts but when number are uncertain so its try to near number return 
sns.histplot(df['Income_log'],kde=True,edgecolor="black")
plt.xlabel(" distribution of income (after transformation)")
plt.title("income number")
plt.ylabel("frequency")
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , LabelEncoder
df = pd.read_csv(r"C:\Users\Lenovo\Desktop\gaurav\tips.csv")
# standard use
df = pd.read_csv(r"C:\Users\Lenovo\Desktop\gaurav\titanic.csv")
label  = LabelEncoder()
df["name_log"]= label.fit_transform(df["Name"])
df["ticket_log"]= label.fit_transform(df["Ticket"])
features = ['name_log', 'ticket_log']
target = ['Age']
X = df[features]
y = df[target]
X.head()
y.head()
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.head()
y_test.head()
plt.scatter(X_train['name_log'],y_train['Age'],c= y_train["Age"],cmap='viridis', alpha=0.5)
plt.title('Before Scaling')
plt.xlabel('Age')
plt.ylabel('EstimatedAge')
```

```python
scaler = StandardScaler()
scaler.fit(X_train[['name_log', 'ticket_log']])
X_train[['name_log', 'ticket_log']] = scaler.transform(X_train[['name_log', 'ticket_log']])
X_test[['name_log', 'ticket_log']] = scaler.transform(X_test[['name_log', 'ticket_log']])
plt.scatter(X_train['name_log'],y_train['Age'],c= y_train["Age"],cmap='viridis', alpha=0.5)
plt.title('after Scaling')
plt.xlabel('Age')
plt.ylabel('EstimatedAge')
```
