```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , LabelEncoder
df = pd.read_csv(r"C:\Users\Lenovo\Desktop\gaurav\tips.csv")
# standard use
label = LabelEncoder()
df['sex_log'] = label.fit_transform(df['sex'])
```

```python
features = ['sex_log', 'tip', 'total_bill']
target = 'size'
X = df[features]
y = df[target]
X.head()
y.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.head()
y_test.head()
```

```python
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 2)
plt.scatter(X_train['sex_log'], X_train['tip'],X_train['total_bill'], c=y_train, cmap='viridis', alpha=0.5)
plt.title('Before Scaling')
plt.xlabel('Age')
plt.ylabel('EstimatedSalary')
```

```python
scaler = StandardScaler()
scaler.fit(X_train[['sex_log', 'tip']])
```

```python
X_train[['sex_log', 'tip','total_bill']] = scaler.transform(X_train[['sex_log', 'tip']])
X_test[['sex_log', 'tip','total_bill']] = scaler.transform(X_test[['sex_log', 'tip']])
```

```python
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 2)
plt.scatter(X_train['sex_log'], X_train['tip'],X_train['total_bill'], c=y_train, cmap='viridis', alpha=0.5)
plt.title('after Scaling')
plt.xlabel('Age')
plt.ylabel('EstimatedSalary')
```
